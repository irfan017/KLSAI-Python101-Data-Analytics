```python
# assigning data as the dataframe with only selected columns
data = df[["Country", 'Region', 'Population', 'Area (sq. mi.)','GDP ($ per capita)']]
```

```python
data.head()
```

```python
# renaming column
data = data.rename(columns= {'Area (sq. mi.)' : 'Area'})
```

```python
# checking wether column is renamed
data.head()
```

```python
# renaming column
data = data.rename(columns= {'GDP ($ per capita)' : 'GDP'})
```

```python
data.head()
```

```python
# determining wether there is a null value in the dataframe
data.isnull().values.any()
```

```python
# checking the total sum of cells with null value
data.isnull().sum()
```

```python
# checking whichi row contains the null value
data[data.isnull().any(axis=1)]
```

```python
# checking the length of the dataframe
len(data)
```

```python
# droping the row with NaN in it and reassigninig it to the data variable
data = data.dropna()
```

```python
# checking wether the row has been dropped
# note that the index number did not change when a row is dropped. it can be seen that number 223 is missing
data.tail()
```

```python
# the len of the dataframe is now less because we dropped a row
len(data)
```

```python
# reset the index of the dataframe
data = data.reset_index(drop=True)
```

```python
# the index has been reset
data.tail()
```

```python
# cleaning the data set 
# below it can be seen that there are unecessary white spaces after the words. this makes it hard to sort the data 
# so we have to remove this 
data['Region'].unique()
```

```python
# reassigning to the dataframe 
data['Region']=data['Region'].str.strip()
```

```python
# now it can be seen that it has no unecessary white space
data['Region'].unique()
```

```python
# plotting graphs
```

```python
# importing relevant packages
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
# grouping the dataframe by region and obtaining only the minimum value of the population in that region
dmin = data.groupby(['Region'], sort=False)['Population'].min()
```

```python
# displaying the minimum value according to the region
dmin
```

```python
# plotting a bar chart of the Mimimum Population per region
dmin.plot.bar()
plt.title('Minimum population per region')
plt.show()
```

```python
# exercise
```

```python
# make a graph of the maximum area per region

dmax = data.groupby(['Region'], sort=False)['Area'].max()
dmax.plot.bar()
plt.title('Maximum area per region')
plt.show()
```
